```python
import seaborn as sns
from matplotlib import pyplot as plt

import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('darkgrid')

sns.pairplot(tips)
```

```python
tips.dtypes
# seaborn is plotting the numeric values in the pairplot
```

```python
# pairplot combines many different seaborn plots, on diagonal we usually see histograms or KDEs and on non diagonals we seesctater plots or regplot
g = sns.pairplot(tips,diag_kind='kde',kind='reg')

# kind refers to non diagonals
```

```python
type(g)

# pair plot returns a pir grid
```

```python
g = sns.pairplot(tips)
g.map_upper(sns.kdeplot,n_levels = 6,color="xkcd:wine red")
```

```python
tips['weekend'] = tips.day.isin(['Sun','Sat'])
tips.head()
```

```python
sns.pairplot(tips,hue='weekend')
```

```python
sns.pairplot(tips,hue='weekend',vars=['total_bill','tip'])
```

```python
sns.pairplot(tips,hue='weekend',y_vars=['tip'],x_vars=['total_bill','tip'],kind='reg',height=12)
```

```python
tips.drop('weekend',axis=1,inplace=True)
```

```python
sns.pairplot(tips,diag_kws={'color':'gray','alpha':0.7})
```
