```python
import seaborn as sns
from matplotlib import pyplot as plt

import numpy as np
import pandas as pd
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.shape
```

```python
sns.set_style('darkgrid')
sns.barplot(x='species',y='body_mass_g',data=penguins)
```

```python
penguins.groupby('species').body_mass_g.mean()
# so the bar plt actually displays the mean of the above dataframe
```

```python
sns.barplot(y='species',x='body_mass_g',data=penguins)

# confidence levels are created here on the basis of bootstrapping process
```

```python
sns.barplot(y='species',x='body_mass_g',data=penguins,errorbar=None)
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,estimator='std')
```

```python
penguins.groupby('species').body_mass_g.describe()
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,hue='sex',order=['Gentoo','Chinstrap','Adelie'])
```
