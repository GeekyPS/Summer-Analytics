# Box Plot

```python
# 25th percentile is till where 25% data has occurred, median is till where 50% has occurred and similarly 75th percentile and then box plot is between these points

# IQR --> Inner Quartile Range, distance between 25th percentile and the 75th percentile
""" whiskers --> stretch the whisker on left hand side to max(25th percentile-1.5*PQR,last datapoint i.e the leftest one),
on the right hand side take minimum of 75th percentile + 1.5pqr and closest datapoint to it on left , the remaining data pont on right if any left is called outlier
"""

# Box plot is also called Categorical Distribution Plot i.e it is usefule when splitting data in categories

```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
# filtered the data
```

```python
sns.boxplot(cars.mpg,orient='h')
```

```python
cars.mpg.describe()
# as visible we have no outliers so minimum value of 9 and maximum 46.6 align with the whiskers 
```

```python
# now we can categorize the data
sns.boxplot(x='origin',y='mpg', data=cars)
```

```python
sns.boxplot(x='origin',y='mpg',hue='cylinders', data=cars)
# this way we can see that the reason usa has terrible mpg is that it is the only of the three which produces 8 cylinder cars
```

```python
cars['newer_model']=cars.model_year>76
```

```python
sns.boxplot(x='origin',y='mpg',hue='newer_model', data=cars)

# as visible newer models have better miles per gallon than older models
```

```python
sns.boxplot(y='origin',x='mpg',hue='newer_model', data=cars,order=['japan','europe','usa'])
```

```python
sns.boxplot(y='origin',x='mpg',hue='newer_model', data=cars,order=['japan','europe','usa'],color='g')
```

```python
sns.boxplot(x='origin',y='mpg', data=cars,whis=1)
# this decreases the times IQR is multiplied to calculate whisker
# black dots are outliers of the data, if we decrease value of whiskers , number of outliers decreases
```

```python
sns.boxplot(x='origin',y='mpg', data=cars,whis=1,fliersize=2)
# fliersize is
```

### Note: Seaborn have even more customization than that given on seaborn docs as it is build on top of matplotlib so it inherits all properties

```python
sns.boxplot(x='origin',y='mpg', data=cars,whis=1,showcaps=False,fliersize=10)
```
