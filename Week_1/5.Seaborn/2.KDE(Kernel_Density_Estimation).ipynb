```python
# this allows to estimate probablity density function from our dataset
# seabord's default is gaussian kernel , and KDE is superpositions of such kernels

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.head()
```

### KDE Univariate plot( one variable plot)

```python
sns.kdeplot(cars.horsepower,fill=True)
```

```python
sns.kdeplot(cars.horsepower,fill=True,bw_adjust=0.1)
# increasing bw_adjust makes the curve smoother , decreasing it makes it sharp but now modes are easily available
```

```python
sns.kdeplot(cars.horsepower,cumulative=True,fill=True)
```

### Bivariate KDE plot

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg)
```

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg,n_levels = 20)
```

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg,fill=True,cbar=True)
```
