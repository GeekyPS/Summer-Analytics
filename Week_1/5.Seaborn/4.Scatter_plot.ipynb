```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds').dropna()
diamonds.shape
```

```python
diamonds.head()
```

```python
sns.set_style('dark')
```

```python
diamonds = diamonds[diamonds.color.isin(['D','F','J'])].sample(n=100,random_state=22)
# here 100 random sample is taken
```

```python
sns.scatterplot(x =diamonds.carat,y = diamonds.price)
```

```python
sns.scatterplot(x='x',y='y',data=diamonds)
```

```python
sns.scatterplot(x='carat',y='price',hue='cut',data=diamonds)
```

```python
sns.scatterplot(x='carat',y='price',hue='depth',data=diamonds)
```

```python
sns.scatterplot(x='carat',y='price',hue='depth',data=diamonds,s=100)
```

```python
sns.scatterplot(x='carat',y='price',data=diamonds,style='cut',s=100)
```

```python
sns.scatterplot(x='carat',y='price',data=diamonds,size='cut',s=100)
```

```python
sns.scatterplot(x='carat',y='price',data=diamonds,s=100,alpha=0.5,marker='*',edgecolor='r')
```
