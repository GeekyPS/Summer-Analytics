```python
import seaborn as sns
from matplotlib import pyplot as plt

import numpy as np
```

```python
cars = sns.load_dataset('mpg')
cars.head()
```

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
cars.groupby('origin').cylinders.value_counts().unstack()
# unstack produces datframe from multindex series
```

```python
cars.groupby('origin').cylinders.value_counts().unstack().fillna(0)
```

```python
origin_cyl = cars.groupby('origin').cylinders.value_counts().unstack().fillna(0)
sns.heatmap(origin_cyl)
```

```python
test_array = np.array([[1,2],[3,4]])
sns.heatmap(test_array)
# heat maps could not only be made on datframe but on numpy arrays as well
```

```python
#this is used to take transpose of the dataframe
origin_cyl.T
```

```python
sns.heatmap(origin_cyl.T)
```

```python
cars.drop('origin', axis=1).drop('name',axis=1).corr()

# this tells the correlation between different values ,by default this is the pearson correlation, name and origin columns where dropped as they were non numeric columns
```

```python
sns.heatmap(cars.drop('origin', axis=1).drop('name',axis=1).corr())
```

```python
sns.heatmap(origin_cyl,cmap='Blues')
```

```python
sns.heatmap(cars.drop('origin', axis=1).drop('name',axis=1).corr(),cmap='RdBu')
```

```python
sns.heatmap(cars.drop('origin', axis=1).drop('name',axis=1).corr(),cmap='RdBu',center=0)
# this sets the white center to 0 value
```

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True)
```

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True,fmt='0.0f')
# format in which annotation appears this specifies we want zero decimal places in code
```

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True,fmt='0.1f')
```

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True,fmt='0.1f',annot_kws={
    'fontsize':16,
    'fontfamily':'serif'
})
```

```python
sns.heatmap(cars.drop('origin', axis=1).drop('name',axis=1).corr(),cmap='RdBu',linewidths=1)
```
