```python
x = np.arange(4) # [0 1 2 3]
plt.xticks(x,['Ankit','Kuldeep','Aryan','Himanshi'])
# assigning names to x axis parameters

y= [8.8,5.2,3.6,5.9]

plt.bar(x,y,color=['g','lime','b','brown'])
plt.show()
```

```python
d = {
    'Red O': np.random.rand(10), # returns a list of 10 random floats in the range [0,1)
    'Grn X': np.random.rand(10)
}

df = pd.DataFrame(d)
print(df)
```

```python
df.plot(style=['ro','gx'])
plt.show()
```

```python
ts = pd.Series(np.random.randn(180),index=pd.date_range('1/1/2018',periods=180))
# randn returns staandard normal distribution and returns list of 180 different values, similarly date range returns 180 different dates starting from starte date and increasing 1 month each
df = pd.DataFrame(np.random.randn(180,3),index=ts.index,columns=list('ABC'))

df.cumsum().plot()
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)

sizes = (30*np.random.rand(50))**2

plt.scatter(x,y,s=sizes,c=colors,alpha=0.5)
# alpha is transparency setting
plt.show()
```

```python
fig = plt.figure()
fig.suptitle('My subplots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222)
plt.plot([np.sin(n) for n in np.arange(0,10,0.01)])
fig.add_subplot(223)
fig.add_subplot(224)

plt.show()
```

```python
fig,plots = plt.subplots(2,sharex=True)
fig.suptitle('Sharing X axis')

x = range(0,200,5)
y = [n**0.8 for n in x]
plots[0].plot(x,y,color='r')
plots[1].scatter(x,y,color='g')
plt.savefig('abc.png')
plt.show()
```
