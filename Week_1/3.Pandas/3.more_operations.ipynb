# Rename, remove, sort, filter series of a dataframe

```python
import pandas as pd

ufo = pd.read_csv('https://bit.ly/uforeports')
ufo.head()
```

### Rename Columns

```python
ufo.columns 
```

```python
ufo.rename(columns={'Colors Reported': 'Colors_Reported','Shape Reported':'Shape_Reported'},inplace=True)
ufo.head()
```

```python
ufo_cols = ['city','colors reported','shape reported','state','time'];
ufo.columns = ufo_cols
ufo.head()
```

```python
ufo = pd.read_csv('https://bit.ly/uforeports',names=ufo_cols,header=0)
# header = 0 means existing tables first row should be treated as header, 
ufo.head()
```

```python
ufo.columns = ufo.columns.str.replace(' ','_')
ufo.head()
```

### Remove data from pandas dataframe


```python
ufo.shape
```

```python
ufo.drop('colors_reported',axis=1,inplace=True)
ufo.head()
# axis 0 is rows and axis 1 is columns , so this tells pandas what we want to remove, The 'inplace=True' argument stands for the data frame has to make changes permanent
```

```python
ufo.drop(['city','state'],axis=1,inplace=True)
ufo.head()
```

```python
ufo.drop([0,1],axis=0,inplace=True)
# removing first two rows
ufo.head()
```

### sort pandas dataframe

```python
movies = pd.read_csv('https://bit.ly/imdbratings')
movies.head()
```

```python
movies.title.sort_values(ascending=False)

# this returns a panda series, by default this is not changing the data permanently
```

```python
movies.sort_values('title',ascending=False)
# this method returns a dataframe
```

### Filter rows of pandas dataframe by column value

```python
movies = pd.read_csv('https://bit.ly/imdbratings')
movies.head()
```

```python
booleans = []
for length in movies.duration:
    if length>=200:
        booleans.append(True)
    else:
        booleans.append(False)    

booleans[0:5]        
```

```python
is_long =  pd.Series(booleans)
is_long.head()
# this converts this array into panda series
```

```python
movies[is_long]
# this will do the filtering
```

```python
is_long = movies.duration >=210 # this is much smaller method
movies[is_long]
```

```python
movies = pd.read_csv('https://bit.ly/imdbratings',usecols=['title','duration','genre'])
# read only ceratin columns from the dataframe

movies[((movies.duration>=215) & (movies.genre=='Drama'))]
# this is the syntax for anding two conditions
```
