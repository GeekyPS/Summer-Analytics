```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
```

```python
mobile = tf.keras.applications.mobilenet_v2.MobileNetV2()
# deep learning model pre trained with weights and biases
```

## Pre Processing the image

```python
from keras.preprocessing import  image
```

```python
img = image.load_img("cat.jpg",target_size=(224,224))
plt.imshow(img)
plt.show()
```

```python
resized_image = image.img_to_array(img)
resized_image.shape
# 3 layers of rgb values in here
```

```python
final_image = np.expand_dims(resized_image,axis=0)
# need 4th dimension
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
final_image.shape
```

```python
predictions = mobile.predict(final_image)
predictions
```

```python
from keras.applications import imagenet_utils
```

```python
results = imagenet_utils.decode_predictions(predictions)
results
```
