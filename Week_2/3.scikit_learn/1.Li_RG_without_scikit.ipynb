## Linear Regression without scikit learn

```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt

penguins = pd.read_csv('penguins_size.csv')
penguins.dropna(inplace=True)
penguins.head()
```

```python
plt.figure(figsize=(10,10))
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")


sns.scatterplot(x="flipper_length_mm",y="body_mass_g",data=penguins,color="blue",alpha=0.7)
```

```python
# in this problem penguin mass is our target , so this is a regression problem not a classification problem , also visibly the relationship between the two are almost linear


# here x is flipper lenth and y is body mass
def linear_model_body_mass(flipper_length,slope,intercept_body_mass):
    expected_body_mass = slope*flipper_length + intercept_body_mass
    return expected_body_mass
```

```python
slope =( penguins.body_mass_g.max() -  penguins.body_mass_g.min())/(penguins.flipper_length_mm.max()- penguins.flipper_length_mm.min())
# prediction of slope by max min values
slope
```

```python

intercept_body_mass = -8000 # just a guess
flipper_length_range = np.linspace(penguins.flipper_length_mm.max(), penguins.flipper_length_mm.min(),num=300)
predicted_body_mass = linear_model_body_mass(flipper_length_range,slope,intercept_body_mass,)
```
